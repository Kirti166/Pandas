3. Low Level Data Understanding :


- a. Find count of unique values in location column.
- b. Find which continent has maximum frequency using values
counts.
- c. Find maximum & mean value in 'total_cases'.
- d. Find 25%,50% & 75% quartile value in 'total_deaths'.
- e. Find which continent has maximum
'human_development_index'.
- f. Find which continent has minimum 'gdp_per_capita'

```python
## a. Find count of unique values in location column.

df['location'].nunique()
```

```python
## b. Find which continent has maximum frequency using values counts.

df['continent'].value_counts().idxmax()
```

```python
## c. Find maximum & mean value in 'total_cases'.

max_total_cases = df['total_cases'].max()
print('max_total_cases:', max_total_cases)
mean_total_cases = df['total_cases'].mean()
print('mean_total_cases:', mean_total_cases)
```

```python
## d. Find 25%,50% & 75% quartile value in 'total_deaths'.

quartiles_total_deaths = df['total_deaths'].quantile([0.25, 0.5, 0.75])
print(quartiles_total_deaths )
```

```python
## e. Find which continent has maximum 'human_development_index'.

max_hdi_continent = df[df['human_development_index'] == df['human_development_index'].max()]['continent'].iloc[0]
max_hdi_continent
```

```python
## f. Find which continent has minimum 'gdp_per_capita'

min_gdp_continent = df[df['gdp_per_capita'] == df['gdp_per_capita'].min()]['continent'].iloc[0]
print(min_gdp_continent)
```

### 5. Data Cleaning

 - a. Remove all duplicates observations
 - b. Find missing values in all columns
 - c. Remove all observations where continent column value is
missing
 Tip : using subset parameter in dropna
 - d. Fill all missing values with 0

```python
## a. Remove all duplicates observations

df = df.drop_duplicates()
df
```

```python
## b. Find missing values in all columns

df.isnull().sum()
```

```python
## c. Remove all observations where continent column value is missing Tip : using subset parameter in dropna

df = df.dropna(subset=['continent'])
df
```

```python
## d. Fill all missing values with 0

df = df.fillna(0)
df
```

### 6. Date time format :
 - a. Convert date column in datetime format using
pandas.to_datetime
 -b. Create new column month after extracting month data from
date
 column.

```python
## a. Convert date column in datetime format using pandas.to_datetime

df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
df['month'] = df['date'].dt.month
df
```

### 7. Data Aggregation:
 - a. Find max value in all columns using groupby function on
'continent'
 column
 Tip: use reset_index() after applying groupby
 - b. Store the result in a new dataframe named 'df_groupby'.
 (Use df_groupby dataframe for all further analysis)


```python
## a. Find max value in all columns using groupby function on 'continent' column Tip: use reset_index() after applying groupby

df.groupby('continent').max().reset_index()
```

```python
### b. Store the result in a new dataframe named 'df_groupby'. (Use df_groupby dataframe for all further analysis)

df_groupby = df.groupby('continent').max().reset_index()
df_groupby
```

### 8. Feature Engineering :
 a. Create a new feature 'total_deaths_to_total_cases' by ratio of
 'total_deaths' column to 'total_cases'

```python
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
df_groupby
```

## 9. Data Visualization :
 - a. Perform Univariate analysis on 'gdp_per_capita' column by
plotting
 histogram using seaborn dist plot.
 - b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'
 - c. Plot Pairplot on df_groupby dataset.
 - d. Plot a bar plot of 'continent' column with 'total_cases' .

```python
## a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot.

sns.displot(df_groupby['gdp_per_capita'], kde=True)
plt.show()
```

```python
## b. Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

sns.scatterplot(x='total_cases', y='gdp_per_capita', data=df_groupby)
plt.show()
```

```python
## c. Plot Pairplot on df_groupby dataset.

numerical_columns = ['total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']


sns.pairplot(df_groupby[numerical_columns])
plt.show()
```

```python
## d. Plot a bar plot of 'continent' column with 'total_cases' . Tip : using kind='bar' in seaborn catplot

sns.catplot(x='continent', y='total_cases', data=df_groupby, kind='bar')
plt.show()
```
